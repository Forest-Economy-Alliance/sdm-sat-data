```python
from pathlib import Path
import pandas as pd
import numpy as np
from functools import reduce
```

```python
pd.set_option('display.max_columns', 1000)
```

```python
# Defining directories
dir_path = Path.cwd().parent
cov_dfs = list(dir_path.joinpath("data/cleaned/covariates_original").glob("*.csv"))
cov_dfs_no_nan = list(dir_path.joinpath("data/cleaned/covariates_smoothened").glob("*.csv"))


prep_folder = dir_path.joinpath("data/prep")

processed_folder = dir_path.joinpath("data/processed")

cov_final_path = processed_folder.joinpath("cov_final.parquet")
```

```python
cov_dfs
```

```python
data_holder = []
for df_path in cov_dfs:
    
    df = pd.read_csv(df_path, dtype={"long": str, "lat": str})
    data_holder.append(df)
```

```python
alt_data_holder = []
for alt_df_path in cov_dfs_no_nan:
    alt_df = pd.read_csv(alt_df_path, dtype={"long": str, "lat": str})
    alt_data_holder.append(alt_df)
```

```python
data_holder[0]
```

```python
merged_df = reduce(lambda left, right: pd.merge(left, right, on=['long', 'lat', 'species_names'], how='outer', validate="1:1"), data_holder)
merged_df
```

```python
alternate_no_nan_cov = reduce(lambda left, right: pd.merge(left, right, on=['long', 'lat', 'species_names'], how='outer', validate="1:1"), alt_data_holder)
alternate_no_nan_cov
```

### Missingness in Covariates

sand, cfvo, phh2o, clay, nitrogen, cec, silt and ocd have 58 missing locations. 

bdod, soc = 17 missing locations

So for all the misisng values in the original Covariates data frame, we will impute the bilinear resampled value for alternate covariates frame.

```python
for col in merged_df.columns:
    print(merged_df[col].isna().value_counts())
```

```python
for col in alternate_no_nan_cov.columns:
    print(alternate_no_nan_cov[col].isna().value_counts())
```

### Imputation of missing values

```python
impt_df = pd.merge(left=merged_df, right=alternate_no_nan_cov, on=['long', 'lat', 'species_names'], how='outer', validate="1:1")
impt_df
```

```python
for col in merged_df.columns:
    if col not in ['long', 'lat', 'species_names', 'elevation', 'slope', 'aspect']:
        alt_col = col + "_250"
        impt_df[col] = np.where(impt_df[col].isna(), impt_df[alt_col], impt_df[col])
```

```python
for col in impt_df.columns:
    print(impt_df[col].isna().value_counts())
```

```python
cov_df_final = impt_df.copy()

for col in cov_df_final:
    if str(col).endswith("_250"):
        cov_df_final.drop(col, axis=1, inplace=True)
cov_df_final
```

```python
for col in cov_df_final.columns:
    print(cov_df_final[col].isna().value_counts())
```

### Scaling

Scaling all all soil columns except nitrogen and bulk density by 10 and the these two by 100. elevation, slope and aspect are not scaled

```python
for col in cov_df_final.columns:
    if col not in ['long', 'lat', 'species_names', 'elevation', 'slope', 'aspect']:

        if col in ["nitrogen_f", "bdod_f"]:
            cov_df_final[col] = cov_df_final[col]/100

        else:
            cov_df_final[col] = cov_df_final[col]/10
```
