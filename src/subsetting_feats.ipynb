## Subsetting Features 

```python
from pathlib import Path
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import json
```

```python
# Defining directories
dir_path = Path.cwd().parent
data_folder = dir_path.joinpath("data/processed")
df_path = data_folder.joinpath("model_df.parquet")

out_path = data_folder.joinpath("subsetted_model_df.parquet")
```

```python
df = pd.read_parquet(df_path)
df
```

Feature subset chosen on the basis of feature importance greater than 0.01

```python
# column_ranges = [
#     range(0, 12),
#     range(44, 58),
#     range(76, 78),
#     range(82, 84),
#     range(68, 70),
#     range(90, 103)
# ]

# Combine all ranges into a single list of column indices
# selected_columns = [col for r in column_ranges for col in r]

cols_to_keep = ["long", "lat", "year", "species"]+['B11summer', 'B11winter', 'B12summer', 'B12winter', 'B08summer',
       'B03summer', 'CVIsummer', 'PVIsummer', 'sand_f', 'elevation', 'cfvo_f',
       'phh2o_f', 'soc_f', 'slope', 'clay_f', 'nitrogen_f', 'cec_f', 'bdod_f',
       'aspect', 'silt_f', 'ocd_f'] 



# Select the specified columns from the DataFrame
subset_df = df[cols_to_keep]
subset_df
```
